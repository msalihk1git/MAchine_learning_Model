```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    learningrate=0.001
    iterations=1000
    n=len(x)

    for i in range(iterations):
        y_predicted=m_curr * x + b_curr
        md=-(2/n)*sum(x*(y-y_predicted))                       #derivative of m
        
        bd=-(2/n)*sum(y-y_predicted)     #derivative of b

        m_curr=m_curr-learningrate *md
        b_curr=b_curr-learningrate* bd

        print ("m {}, b {},  iteration {}".format(m_curr,b_curr, i))
```

```python
gradient_descent(x,y)
```

```python
gradient_descent(x,y)    #learning rate=0.001
```
