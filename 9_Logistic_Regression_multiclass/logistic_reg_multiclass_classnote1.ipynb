```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data=pd.read_csv("winequality.csv",sep=";")
```

```python
corrmatrix=data.corr()['quality']
```

```python
corrmatrix
```

```python
corrmatrix[abs(corrmatrix)<0.10].index
```

```python
lowcorr=corrmatrix[abs(corrmatrix)<0.10].index
```

```python
data.isna().sum()
```

```python
y=data['quality']
X=data.drop('quality',axis=1)
X.drop(lowcorr,axis=1,inplace=True)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
lr_classifier=LogisticRegression()
```

```python
lr_classifier.fit(X_train,y_train)
```

```python
lr_classifier.score(X_train,y_train)
```

```python
lr_classifier.score(X_test, y_test)
```
